## Masking

```python
img = cv.imread('traffic_sign.ppm')
display(img)
```

```python
# mask = np.zeros(shape=img.shape, dtype=img.dtype)    
mask = np.zeros_like(img)

print(mask.shape)
print(mask.dtype)
```

```python
plt.imshow(mask)
```

```python
mask[40:100, 50:100] = 255     # Broadcast along the 3 dimensions
plt.imshow(mask)
```

```python
mask[60, 60]   
```

```python
masked_img = np.bitwise_and(img, mask)
```

```python
# Triangle mask
img = cv.imread('traffic_sign.ppm')
mask = np.zeros_like(img)

# Vertices has (x,y) format (you are dealing with opencv function)
vertices = np.array([[50, 100], [75, 40], [100,100]], dtype=np.int32)
cv.fillPoly(mask, [vertices], (255,255,255))  # Modifies in place and returns a copy

plt.imshow(mask)
```

```python
display(np.bitwise_and(img, mask))
```
